```python
#Kaggle
print("Kaggle API kuruluyor...")
!pip install kaggle -q


from google.colab import files
import os
print("Kaggle.json dosyanızı yükleyin:")
uploaded = files.upload()

!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
print("Veri seti indiriliyor...")
!rm -rf /content/gtsrb-german-traffic-sign*
!rm -rf /content/Train /content/Test /content/Meta.csv

!kaggle datasets download -d meowmeowmeowmeowmeow/gtsrb-german-traffic-sign --force
!unzip -o -q gtsrb-german-traffic-sign.zip

print("Veri seti başarıyla indirildi!")
!ls -la /content/

# Kütüphalener
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from PIL import Image
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tqdm import tqdm
import warnings
import seaborn as sns

warnings.filterwarnings('ignore')


# GPU kullanmı için runtime türü GPU seçilmelidir
# GPU Yapılandırması
print("GPU Yapılandırması başlatılıyor...")
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"{len(gpus)} GPU bulundu ve yapılandırıldı")
    except RuntimeError as e:
        print(f"GPU yapılandırma hatası: {e}")
else:
    print("GPU bulunamadı, CPU kullanılacak")

# Global strateji
strategy = tf.distribute.MirroredStrategy()
print(f"Dağıtım stratejisi: {strategy.num_replicas_in_sync} cihaz")

##


print(f"TensorFlow Version: {tf.__version__}")
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")

# Traffic sign class names (43 classes)
class_names = {
    0: 'Speed limit (20km/h)',
    1: 'Speed limit (30km/h)',
    2: 'Speed limit (50km/h)',
    3: 'Speed limit (60km/h)',
    4: 'Speed limit (70km/h)',
    5: 'Speed limit (80km/h)',
    6: 'End of speed limit (80km/h)',
    7: 'Speed limit (100km/h)',
    8: 'Speed limit (120km/h)',
    9: 'No passing',
    10: 'No passing for vehicles over 3.5 metric tons',
    11: 'Right-of-way at the next intersection',
    12: 'Priority road',
    13: 'Yield',
    14: 'Stop',
    15: 'No vehicles',
    16: 'Vehicles over 3.5 metric tons prohibited',
    17: 'No entry',
    18: 'General caution',
    19: 'Dangerous curve to the left',
    20: 'Dangerous curve to the right',
    21: 'Double curve',
    22: 'Bumpy road',
    23: 'Slippery road',
    24: 'Road narrows on the right',
    25: 'Road work',
    26: 'Traffic signals',
    27: 'Pedestrians',
    28: 'Children crossing',
    29: 'Bicycles crossing',
    30: 'Beware of ice/snow',
    31: 'Wild animals crossing',
    32: 'End of all speed and passing limits',
    33: 'Turn right ahead',
    34: 'Turn left ahead',
    35: 'Ahead only',
    36: 'Go straight or right',
    37: 'Go straight or left',
    38: 'Keep right',
    39: 'Keep left',
    40: 'Roundabout mandatory',
    41: 'End of no passing',
    42: 'End of no passing by vehicles over 3.5 metric tons'
}

# Veri yükleme fonksiyonu
def load_data():
    data = []
    labels = []
    classes = 43

    print("Eğitim verileri yükleniyor...")
    for i in tqdm(range(classes), desc="Classes"):
        path = os.path.join('/content/Train', str(i))
        if os.path.exists(path):
            images = os.listdir(path)

            for img_name in images:
                try:
                    img_path = os.path.join(path, img_name)
                    image = Image.open(img_path)
                    image = image.resize((32, 32))  # Increased size for better performance
                    image = np.array(image)

                    # Ensure RGB format
                    if len(image.shape) == 3 and image.shape[2] == 3:
                        data.append(image)
                        labels.append(i)
                except Exception as e:
                    print(f"Error loading image {img_name}: {e}")
        else:
            print(f"Path not found: {path}")

    return np.array(data), np.array(labels)

# Veri setini yükle
X_data, y_data = load_data()

print(f"Loaded data shape: {X_data.shape}")
print(f"Loaded labels shape: {y_data.shape}")
print(f"Number of classes: {len(np.unique(y_data))}")

# Veri setini görselleştir
def visualize_data(X, y, num_samples=25):
    plt.figure(figsize=(12, 12))
    for i in range(min(num_samples, len(X))):
        plt.subplot(5, 5, i + 1)
        plt.imshow(X[i])
        plt.title(f'Class: {y[i]}')
        plt.axis('off')
    plt.suptitle('Sample Traffic Signs', fontsize=16)
    plt.tight_layout()
    plt.show()

# Sınıf dağılımını görselleştir
def plot_class_distribution(y):
    plt.figure(figsize=(15, 6))
    unique, counts = np.unique(y, return_counts=True)
    plt.bar(unique, counts)
    plt.xlabel('Class')
    plt.ylabel('Number of Samples')
    plt.title('Class Distribution')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Veri görselleştirmesi
visualize_data(X_data, y_data)
plot_class_distribution(y_data)

# Veri ön işleme
def preprocess_data(X, y):
    # Normalizasyon
    X = X.astype('float32') / 255.0

    # One-hot encoding
    y = to_categorical(y, 43)

    return X, y

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.2, random_state=42, stratify=y_data
)

print(f"Train set: {X_train.shape}, {y_train.shape}")
print(f"Test set: {X_test.shape}, {y_test.shape}")

# Veri ön işleme
X_train, y_train = preprocess_data(X_train, y_train)
X_test, y_test = preprocess_data(X_test, y_test)

print(f"After preprocessing:")
print(f"Train set: {X_train.shape}, {y_train.shape}")
print(f"Test set: {X_test.shape}, {y_test.shape}")

# Gelişmiş model mimarisi
def create_model():
    model = Sequential([
        # First Convolutional Block
        Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
        BatchNormalization(),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Second Convolutional Block
        Conv2D(64, (3, 3), activation='relu'),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Third Convolutional Block
        Conv2D(128, (3, 3), activation='relu'),
        BatchNormalization(),
        Dropout(0.25),

        # Fully Connected Layers
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(43, activation='softmax')
    ])

    return model

# Model oluştur
model = create_model()

# Model derle
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Model özetini göster
model.summary()

# Callbacks
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=0.0001,
    verbose=1
)

# MODEL EĞİTİMİ - Burada yapılıyor!
print("=" * 50)
print("MODEL EĞİTİMİ BAŞLIYOR!")
print("=" * 50)

history = model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=50,  # Daha fazla epoch
    validation_data=(X_test, y_test),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print("=" * 50)
print("MODEL EĞİTİMİ TAMAMLANDI!")
print("=" * 50)

# Model kaydet
model.save('/content/traffic_signs_model.h5')
print("Model kaydedildi: /content/traffic_signs_model.h5")

# Eğitim geçmişini görselleştir - ACCURACY VE LOSS GRAFİKLERİ
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Accuracy plot
    ax1.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    ax1.set_title('Model Accuracy', fontsize=14)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Loss plot
    ax2.plot(history.history['loss'], label='Training Loss', linewidth=2)
    ax2.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    ax2.set_title('Model Loss', fontsize=14)
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# Grafikleri göster
plot_training_history(history)

# Model değerlendirmesi
print("Model değerlendiriliyor...")
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# Confusion Matrix
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_true_classes, y_pred_classes))

# ÇIKARIM (INFERENCE) FONKSİYONU - İstediğiniz fonksiyon burada!
def predict_traffic_sign(image_path, model):
    """
    Tek bir görüntü için traffic sign prediction yapar

    Args:
        image_path: Görüntü dosyasının yolu
        model: Eğitilmiş model

    Returns:
        predicted_class: Tahmin edilen sınıf
        class_name: Sınıf ismi
        confidence: Tahmin güven skoru
    """
    try:
        # Görüntüyü yükle ve işle
        image = Image.open(image_path)
        image = image.resize((32, 32))
        image = np.array(image)

        # RGB formatına çevir
        if len(image.shape) == 3 and image.shape[2] == 3:
            # Normalizasyon
            image = image.astype('float32') / 255.0

            # Batch dimension ekle
            image = np.expand_dims(image, axis=0)

            # Prediction
            predictions = model.predict(image, verbose=0)
            predicted_class = np.argmax(predictions[0])
            confidence = np.max(predictions[0])

            class_name = class_names.get(predicted_class, "Unknown")

            return predicted_class, class_name, confidence
        else:
            return None, "Error: Invalid image format", 0.0

    except Exception as e:
        return None, f"Error: {str(e)}", 0.0

# Test görüntüleri ile çıkarım testi
def test_inference_with_samples():
    """Test setinden rastgele örnekler ile inference testi"""
    print("=" * 60)
    print("INFERENCE FONKSİYONU TEST EDİLİYOR")
    print("=" * 60)

    # Test setinden rastgele 5 örnek seç
    sample_indices = np.random.choice(len(X_test), 5, replace=False)

    plt.figure(figsize=(15, 10))

    for i, idx in enumerate(sample_indices):
        # Görüntüyü kaydet
        test_image = (X_test[idx] * 255).astype(np.uint8)
        temp_path = f'/content/temp_test_{i}.png'
        Image.fromarray(test_image).save(temp_path)

        # Inference yap
        pred_class, pred_name, confidence = predict_traffic_sign(temp_path, model)
        true_class = np.argmax(y_test[idx])
        true_name = class_names.get(true_class, "Unknown")

        # Görselleştir
        plt.subplot(2, 3, i + 1)
        plt.imshow(test_image)
        plt.title(f'True: {true_name}\nPred: {pred_name}\nConf: {confidence:.3f}',
                 fontsize=10)
        plt.axis('off')

        # Geçici dosyayı sil
        os.remove(temp_path)

        print(f"Sample {i+1}:")
        print(f"  True Class: {true_class} ({true_name})")
        print(f"  Predicted Class: {pred_class} ({pred_name})")
        print(f"  Confidence: {confidence:.4f}")
        print(f"  Correct: {'✓' if pred_class == true_class else '✗'}")
        print()

    plt.tight_layout()
    plt.show()

# Inference testini çalıştır
test_inference_with_samples()

# Dış test verisi ile değerlendirme (varsa)
test_csv_path = '/content/Test.csv'
if os.path.exists(test_csv_path):
    print("=" * 60)
    print("DIŞ TEST VERİSİ İLE DEĞERLENDIRME")
    print("=" * 60)

    test_df = pd.read_csv(test_csv_path)
    print(f"External test data: {len(test_df)} samples")

    correct_predictions = 0
    total_predictions = 0

    for idx, row in tqdm(test_df.iterrows(), total=len(test_df), desc="Testing"):
        image_path = '/content/' + row['Path']
        true_class = row['ClassId']

        if os.path.exists(image_path):
            pred_class, pred_name, confidence = predict_traffic_sign(image_path, model)

            if pred_class is not None:
                total_predictions += 1
                if pred_class == true_class:
                    correct_predictions += 1

    if total_predictions > 0:
        external_accuracy = correct_predictions / total_predictions
        print(f"External Test Accuracy: {external_accuracy:.4f}")
        print(f"Correct: {correct_predictions}/{total_predictions}")

# Sonuç özeti
print("=" * 60)
print("SONUÇ ÖZETİ")
print("=" * 60)
print(f"✓ Veri seti başarıyla yüklendi: {len(X_data)} görüntü")
print(f"✓ Model başarıyla eğitildi")
print(f"✓ Test Accuracy: {test_accuracy:.4f}")
print(f"✓ Model kaydedildi: /content/traffic_signs_model.h5")
print(f"✓ Inference fonksiyonu hazır: predict_traffic_sign()")
print(f"✓ Accuracy ve Loss grafikleri gösterildi")
print("=" * 60)
```
